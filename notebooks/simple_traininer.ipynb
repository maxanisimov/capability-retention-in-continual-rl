```python
import sys
import os
import torch
import torch.nn as nn

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
from src.trainer.SimpleTrainer import SimpleTrainer
from src.data_utils import get_mnist_tasks, _extract_targets
from src.utils import set_random_seed
```

```python
set_random_seed(42)
```

```python
train_tasks, val_tasks, test_tasks = get_mnist_tasks()

model = nn.Sequential()
model.append(nn.Flatten())
model.append(nn.Linear(28 * 28, 128))
model.append(nn.ReLU())
model.append(nn.Linear(128, 10))
```

```python
print(
    f"Tasks: {[torch._unique(_extract_targets(train))[0].tolist() for train in train_tasks]}"
)

trainer = SimpleTrainer(model)
for i, (train, val, _) in enumerate(zip(train_tasks, val_tasks, test_tasks)):
    print(f"Training on task: {i + 1}")
    trainer.train(train, val, epochs=5, batch_size=64)
```

```python
trainer.test(test_tasks)
```
